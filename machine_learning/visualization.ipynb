```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('sample_sensor_data.csv')
df['|a|'] = np.sqrt(df['ax']**2 + df['ay']**2 + df['az']**2)

acceleration_labels = ['ax', 'ay', 'az', '|a|']
acceleration_titles = ['Acceleration X', 'Acceleration Y', 'Acceleration Z', 'Acceleration Magnitude']
gyroscope_labels = ['gx', 'gy', 'gz']
gyroscope_titles = ['Gyroscope X', 'Gyroscope Y', 'Gyroscope Z']
```

```python
fig, axes = plt.subplots(7, 1, figsize=(12, 16))

for i, (label, title) in enumerate(zip(acceleration_labels, acceleration_titles)):
    axes[i].plot(df['ms'], df[label], label=title, color='rgbk'[i])
    axes[i].set_title(title)
    axes[i].set_xlabel('Milliseconds')
    axes[i].set_ylabel('Acceleration (mg)' if i < 4 else 'Gyroscope (mdps)')
    axes[i].legend()
    axes[i].grid(True)

for i, (label, title) in enumerate(zip(gyroscope_labels, gyroscope_titles), start=4):
    axes[i].plot(df['ms'], df[label], label=title, color='rgb'[i-4])
    axes[i].set_title(title)
    axes[i].set_xlabel('Milliseconds')
    axes[i].set_ylabel('Gyroscope (mdps)')
    axes[i].legend()
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```
